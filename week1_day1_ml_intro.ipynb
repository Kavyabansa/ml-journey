# Week 1, Day 1 — ML Intro: Train/Test Split & Bias-Variance

**Goal:** Understand what train/test split and cross-validation mean
before using any library. Build everything from scratch first.

```python
import random

# Step 1: Create 20 random numbers between 1 and 100
random.seed(42)  # seed means you get the same "random" numbers every time you run
data = [random.randint(1, 100) for _ in range(20)] #randint chossing the random number from 1 to 100 1  time but for loop is helping choosing random number 20 times 
print("Full data:", data)
print("Total items:", len(data))

# Step 2: Calculate the split point
# 80% of 20 = 16 items for training, 4 items for testing
split_point = int(len(data) * 0.8)
print("\nSplit point (index):", split_point)

# Step 3: Slice the list at that point
train = data[:split_point]   # items 0 to 15 (16 items)
test  = data[split_point:]   # items 16 to 19 (4 items)

print("\nTraining data:", train)
print("Training size:", len(train))
print("\nTest data:", test)
print("Test size:", len(test))
```

#Cross validation means: split your data into 5 equal chunks, use 4 for training and 1 for testing, rotate which chunk is the test set, repeat 5 times. Here's that logic written out:

```python
# We have 20 data points, split into 5 folds of 4 items each
data = list(range(1, 21))  # [1, 2, 3, ..., 20]
k = 5
fold_size = len(data) // k  # 20 // 5 = 4 items per fold and // is called floor division or integer division. It gives integer per division

print(f"Data has {len(data)} items → {k} folds of {fold_size} each\n")

for fold in range(k):
    # Calculate start and end index for the TEST fold
    start = fold * fold_size       # fold 0→0, fold 1→4, fold 2→8 ...
    end   = start + fold_size      # fold 0→4, fold 1→8, fold 2→12 ...

    # Test fold = this chunk
    test_fold  = data[start:end]

    # Train fold = everything else (before + after the test chunk)
    train_fold = data[:start] + data[end:]

    print(f"Fold {fold+1}:")
    print(f"  TEST  indices {start}–{end-1}: {test_fold}")
    print(f"  TRAIN ({len(train_fold)} items): {train_fold}")
    print()
```

## Bias vs Variance — My Understanding

**Bias** = the model is too simple. It makes wrong assumptions.
Example: fitting a straight line to data that curves. It will be
wrong on training data AND on new data.

**Variance** = the model is too complex. It memorises the training
data including the noise. It scores great on training data but
fails on new data.

**The tradeoff:** making a model more complex reduces bias but
increases variance. The goal is to find the sweet spot in the middle.

Bias = Model bahut simple hai. Training data pe bhi galat, testing pe bhi galat. High bias = underfit.

Variance = Model bahut complex hai. Training data pe bahut accha (low training SS), testing data pe bahut bura (high testing SS). High variance = overfit.

Best model = Low bias + Low variance. Consistent bhi, accurate bhi.
