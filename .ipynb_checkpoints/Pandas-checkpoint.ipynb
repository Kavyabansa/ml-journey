# Single brackets [] → gives you a Series
# Double brackets [[]] → gives you a DataFrame

```python
import pandas as pd
```

# → Series   — use for quick analysis
# → DataFrame — use when passing to sklearn

# loc uses the LABEL — finds row where index = 10
# iloc uses the POSITION — finds first row regardless of label

# iloc uses the Python stdlib indexing scheme, where the first element of the range is included and the last one excluded. So 0:10 will select entries 0,...,9. 
# loc, meanwhile, indexes inclusively. So 0:10 will select entries 0,...,10.

```python
df = pd.DataFrame({
    'name': ['Kavya', 'Rohit', 'Priya'],
    'age':  [20, 22, 21]
})

print(df.shape)
print(df.dtypes)
print(len(df))
```

```python
df = pd.DataFrame({
    'name': ['Kavya', 'Rohit', 'Priya'],
    'age':  [20, 22, 21]
})

print(df['name'])     # single brackets
print(type(df['name']))

print(df[['name']])   # double brackets
print(type(df[['name']]))
```

```python
df = pd.DataFrame({
    'name': ['Kavya', 'Rohit', 'Priya'],
    'age':  [20, 22, 21]
}, index=[10, 20, 30])   # custom index — not 0,1,2
```

```python
df.loc[10]    # loc uses the LABEL — finds row where index = 10
   # iloc uses the POSITION — finds first row regardless of label
```

```python
df.iloc[0] 
```

# apply() = "run this function on each row or each column — your choice — and you can access as many columns as you need inside the function."

# map() = "for each value in this column, do this thing."

# df.dropna()                        # drop rows with any null
# df.dropna(axis=1)                  # drop columns with any null
# df.dropna(how='all')               # drop rows where all values null
# df.dropna(thresh=2)                # keep rows with min 2 non-null
# df.dropna(subset=['age'])          # drop rows where age is null only
# df.dropna(inplace=True)            # modify df directly

# dropna  → use when missing data is rare (<5% of rows)
          and you have enough data to afford losing rows

# fillna  → use when missing data is common (>5%)
          and you can't afford to lose rows

# dropna(axis=1) → use when a whole column is mostly null (>40% missing)
                 that column probably has no useful signal anyway

# how='inner'  # only rows where id exists in BOTH tables
             # safest, most common — no nulls created

# how='left'   # keep ALL rows from df1
             # if no match in df2 → fill with NaN
             # use when df1 is your main table

# how='right'  # keep ALL rows from df2
             # if no match in df1 → fill with NaN

# how='outer'  # keep ALL rows from BOTH tables
             # fills NaN wherever match is missing

df1:                df2:
id  name            id  score
1   A               2   85
2   B               3   90
3   C               4   78

inner → id 2,3 only    → 2 rows
left  → id 1,2,3       → 3 rows (id 1 gets NaN score)
right → id 2,3,4       → 3 rows (id 4 gets NaN name)
outer → id 1,2,3,4     → 4 rows (NaN wherever no match)

# like combining January + February sales data

# axis=1 — stack horizontally (add more columns)
# pd.concat([df1, df2], axis=1)
# use when both tables have same rows
