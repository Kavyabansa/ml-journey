### Task 3 — Confusion Matrix from Scratch
Given y_true and y_pred arrays, calculated:
- TP, TN, FP, FN using a loop with conditions
- Precision = TP / (TP + FP)
- Recall = TP / (TP + FN)

**Key learning:** Precision tells you how many of your
positive predictions were actually correct.
Recall tells you how many actual positives you caught.

```python
import numpy as np

data = np.random.rand(5, 4)

print("Array:\n", data)
print("\nMean of each column (axis=0):", data.mean(axis=0))
print("Max of each row (axis=1):", data.max(axis=1))
print("Std of whole array:", data.std())
```

```python
import numpy as np

arr = np.random.randint(0, 100, 50)
print("Original array:", arr)

above_60 = arr[arr > 60]
print("\nValues above 60:", above_60)

count = above_60.size
print("Count above 60:", count)

arr[arr < 40] = 0
print("\nArray after replacing below 40 with 0:", arr)
```

```python
import numpy as np
y_true = np.array([1, 0, 1, 1, 0, 1, 0, 0, 1, 0])
y_pred = np.array([1, 0, 1, 0, 0, 1, 1, 0, 1, 1])

TP = np.sum((y_pred == 1) & (y_true == 1))
TN = np.sum((y_pred == 0) & (y_true == 0))
FP = np.sum((y_pred == 1) & (y_true == 0))
FN = np.sum((y_pred == 0) & (y_true == 1))

Precision = TP / (TP + FP)
Recall    = TP / (TP + FN)

print("=" * 30)
print("    CONFUSION MATRIX RESULTS")
print("=" * 30)
print(f"  True Positive  (TP) : {TP}")
print(f"  True Negative  (TN) : {TN}")
print(f"  False Positive (FP) : {FP}")
print(f"  False Negative (FN) : {FN}")
print("-" * 30)
print(f"  Precision           : {Precision:.2f}")
print(f"  Recall              : {Recall:.2f}")
print("=" * 30)
```

```python
# Simpler Version
y_true = [1, 0, 1, 1, 0, 1, 0, 0, 1, 0]
y_pred = [1, 0, 1, 0, 0, 1, 1, 0, 1, 1]

TP = 0
TN = 0
FP = 0
FN = 0

for i in range(10):
    if y_pred[i] == 1 and y_true[i] == 1:
        TP += 1
    elif y_pred[i] == 0 and y_true[i] == 0:
        TN += 1
    elif y_pred[i] == 0 and y_true[i] == 1:
        FN += 1
    else:
        FP += 1

Precision = TP / (TP + FP)
Recall    = TP / (TP + FN)

print("=" * 30)
print("    CONFUSION MATRIX RESULTS")
print("=" * 30)
print(f"  True Positive  (TP) : {TP}")
print(f"  True Negative  (TN) : {TN}")
print(f"  False Positive (FP) : {FP}")
print(f"  False Negative (FN) : {FN}")
print("-" * 30)
print(f"  Precision           : {Precision:.2f}")
print(f"  Recall              : {Recall:.2f}")
print("=" * 30)
```
